```python
import numpy as np
import pandas as pd
```

1) Create a Numpy array of size 4 x 5.

```python
narray = np.eye(4,5)
print(narray)
```

2) Randomly initialize the array.

```python
narray = np.random.randint(5,size=(4,5))
print(narray)
```

3) Get the Determinant and Transpose of the Matrix that you created.

```python
print(narray.T) #transpose is 5x4 array
```

4) Create another matrix of size 5 x 4 and randomly initialize it.

```python
matrix1 = narray
matrix2 = np.random.randint(4,size=(5,4))
print(matrix2)
```

5) Perform Matrix multiplication.

```python
np.matmul(matrix1,matrix2)
```

6) Perform element wise matrix multiplication.

```python
matrix1 = np.random.randint(5,size=(3,3))
print ("matrix1: \n",matrix1)
matrix2 = np.random.randint(3,size=(3,3))
print ("matrix2: \n",matrix2)
ans = np.multiply(matrix1,matrix2)
print("ans:\n",ans)
```

7) Find mean, median of the numpy array created.

```python
print("array is :\n",ans)
print("mean is: ",np.mean(ans))
print("median is: ",np.median(ans))
```

8) Obtain each row in the second column of the first array.

```python
print("array: \n",ans)
print(ans[:,1:2])
```
