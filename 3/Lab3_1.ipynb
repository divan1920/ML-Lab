```python
from sklearn import preprocessing
import pandas as pd
from sklearn.naive_bayes import GaussianNB, MultinomialNB
```

```python
le = preprocessing.LabelEncoder()
```

```python
data = pd.read_csv('drive/MyDrive/Dataset1.csv')
data.head()
```

```python
#Lable Encoding each feature 

outlook_encoded = le.fit_transform(data.Outlook)
temp_encoded = le.fit_transform(data.Temp)
humidity_encoded = le.fit_transform(data.Humidity)
wind_encoded = le.fit_transform(data.Wind)
label = le.fit_transform(data.Play)
print(outlook_encoded,temp_encoded,humidity_encoded,wind_encoded,label)
```

```python
#combining all features to make a dataset

features = tuple(zip(outlook_encoded,temp_encoded,humidity_encoded,wind_encoded))
print('Features:',features)
```

```python
from sklearn.model_selection import train_test_split

#split data into training and testing sets

data_train , data_test , target_train , target_test = train_test_split(features,label,test_size=0.2,random_state=35)
```

```python
#train the model using training sets
gnb = GaussianNB()

gnb.fit(data_train,target_train)

```

```python
#prection using model

target_pred = gnb.predict(data_test)
```

```python
from sklearn import metrics

print('Accuracy: ',metrics.accuracy_score(target_test,target_pred))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(target_test,target_pred)
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
precision = precision_score(target_test, target_pred)
recall = recall_score(target_test, target_pred)
print('precision: {}'.format(precision))
print('recall: {}'.format(recall))
```

```python
#train multinomial model using training datasets
mnb = MultinomialNB()

mnb.fit(data_train,target_train)
```

```python
#prediction of target using testing datasets 

target_pred_mnb = mnb.predict(data_test)
```

```python
#how accurate our model is:

print("Accuracy: ",metrics.accuracy_score(target_test,target_pred_mnb))
```

```python
#confusion matrix for mnb model
print("confusion Matrix : \n",confusion_matrix(target_test,target_pred_mnb))
```

```python
precision = precision_score(target_test, target_pred_mnb)
recall = recall_score(target_test, target_pred_mnb)
print('precision: {}'.format(precision))
print('recall: {}'.format(recall))
```
