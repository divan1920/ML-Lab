Aim: Implement Linear Regression Algorithm on the given dataset

```python
import numpy as np
import torch
```

```python
# Input (temp, rainfall, humidity)
inputs = np.array([[73, 67, 43], [91, 88, 64], [87, 134, 58], [102, 43, 37],
                   [69, 96, 70], [73, 67, 43], [91, 88, 64], [87, 134, 58], [102, 43, 37], 
                   [69,96, 70], [73, 67, 43], [91, 88, 64], [87, 134, 58], [102, 43, 37], [69, 96,70]], 
                  dtype='float32')
```

```python
# Targets (apples, oranges)
targets = np.array([[56, 70], [81, 101], [119, 133], [22, 37], [103, 119],
[56, 70], [81, 101], [119, 133], [22, 37], [103, 119],
[56, 70], [81, 101], [119, 133], [22, 37], [103, 119]],
dtype='float32')
```

```python
inputs = torch.from_numpy(inputs)
targets = torch.from_numpy(targets)
```

```python
weight = torch.randn(2, 3, requires_grad=True)
bias = torch.randn(2, requires_grad=True)
```

```python
#define model
def model(inputs):
    return inputs @ weight.t() + bias
#Generate Predictions
preds = model(inputs)
print(preds)
```

```python
#mse loss fuction
def mse(t1, t2):
    diff = t1 - t2
    return torch.sum(diff * diff) / diff.numel()
```

```python
#calculate loss
loss = mse(preds,targets)
print(loss)
```

```python
#compute gradients
loss.backward()
print(weight.grad)
print(bias.grad)
```

#Train the model using gradient descent

```python
with torch.no_grad():
    weight -= weight.grad * 1e-5
    bias -= bias.grad * 1e-5
    weight.grad.zero_()
    bias.grad.zero_()
```

```python
prediction = model(inputs)
loss = mse(prediction, targets)
print(loss)
```

```python
for i in range(100):
    prediction = model(inputs)
    loss = mse(prediction, targets)
    loss.backward()
    with torch.no_grad():
        weight -= weight.grad * 1e-5
        bias -= bias.grad * 1e-5
        weight.grad.zero_()
        bias.grad.zero_()
```

```python
prediction = model(inputs)
loss = mse(prediction, targets)
print(loss)
```
