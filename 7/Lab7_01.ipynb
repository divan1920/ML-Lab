```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
trans_data = sc.fit_transform(data)
```

```python
X = trans_data[:,0:2]
y = trans_data[:,2]
```

```python
y = [0 if a<0 else 1 for a in y]
y = np.array(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=35)
```

```python
class LogisticRegression:
  def sigmoid(self,X):
    return 1/(1+np.exp(-((X).dot(self.Weights)+self.Bias)))
  
  def __init__(self,lr,epochs=100):
    self.learning_rate = lr
    self.epochs = epochs
    pass
  
  def get_loss(self):
    return (np.sum(-1*(self.Y)*np.log(self.sigmoid(self.X))-(1-self.Y)*np.log(1-self.sigmoid(self.X)))/self.Num_samples)
  
  def fit(self,X,Y):
    
    # Assigning data to X and Y
    self.X = X
    self.Y = Y

    # Computing M -> No of training samples 
    # Computing N -> No of features
    self.Num_samples,self.Num_features = X.shape
    print(self.Num_samples,self.Num_features)

    # Initializing all the weights
    self.Weights = np.zeros(self.Num_features)
    self.Bias = 0
    for i in range(self.epochs):
      print("Epoch ",i)
      self.run_optimization()
      print(self.get_loss())
    return self
  
  # Return predictions based on X
  def predict(self,X):
    ret_val = []
    for dat in self.sigmoid(X):
      ret_val.append(1 if dat>=0.5 else 0)
    return ret_val


  def run_optimization(self):
    # Gradient Calculation
    ans = self.sigmoid(self.X)
    t = ( ans - self.Y.T )
    t = np.reshape( t, self.Num_samples ) 

    dW = np.dot( self.X.T, t ) / self.Num_samples
    db = np.sum( t ) / self.Num_samples
    

    self.Weights = self.Weights - self.learning_rate * dW    
    self.Bias = self.Bias - self.learning_rate * db
```

```python
lr = LogisticRegression(0.4,300)
lr.fit(X_train,y_train)
```

```python
lr.Weights
```

```python
def getScore(y,y_pred):
  t = 0 
  f = 0
  for i in range(len(y_pred)):
    if y_pred[i]==y[i]:
      t = t + 1
    else:
      f = f  + 1
  return t/(t+f)
```

```python
y_pred = lr.predict(X_test)
```

```python
getScore(y_test,y_pred)
```
