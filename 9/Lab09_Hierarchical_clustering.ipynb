```python
from sklearn import datasets
from matplotlib import pyplot as plt
import numpy as np
import seaborn as sns
sns.set_style("darkgrid")
```

#Using Iris Dataset

```python
iris = datasets.load_iris()
X = iris['data']
y = iris['target']
```

```python
print("target names: \n",iris.target_names)
print("feature names: \n",iris.feature_names)
```

```python
#rescaling data to a standard deviation of 1
# Import the whiten function
from scipy.cluster.vq import whiten
scaled_data = whiten(X)
```

```python
sns.scatterplot(x=scaled_data[:,0], y=scaled_data[:,1], c =["green"])
```

```python
# Use the linkage() function
distance_matrix = linkage(scaled_data, method = 'ward', metric = 'euclidean')
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
import plotly.figure_factory as ff
fig = ff.create_dendrogram(distance_matrix,)
fig.update_layout(width=800, height=500)
fig.show()
```

###From the dendrogram we can realize that the number of Clusters is 3 and that 2 clusters are closer (the red ones) compared to the green one.

```python
from sklearn.cluster import AgglomerativeClustering
hc_model = AgglomerativeClustering(n_clusters=3, affinity='euclidean',linkage='ward')
hc_model.fit_predict(scaled_data)
```

```python
#pring the data points
plt.scatter(x=scaled_data[:,0], y=scaled_data[:,1], c= hc_model.labels_, cmap='rainbow' )
```

```python
plt.scatter(x=scaled_data[:,0], y=scaled_data[:,1], c=y , cmap='rainbow' )
```

###From those two graphs above we can say that our model performed well as both the the actual and predicted target is similar
